```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Données des deux tableaux combinées
data_combined = {
    'Graphe': ['Polyarbre', 'Faiblement connectés', 'Moyennement connecté', 'Fortement connecté',
               'Polyarbre', 'Faiblement connectés', 'Moyennement connecté', 'Fortement connecté'],
    'Temps d\'inférence (ms)': [0, 27.459, 20.417, 33.943, 2, 2, 1, 2],
    'Temps de propagation (s)': [0, 0.027459, 0.020417, 0.033943, 5, 17, 14, 21],
    'Nombre d\'évidences': [1, 1, 1, 1, 2, 2, 2, 2]
}

# Création du DataFrame
df_combined = pd.DataFrame(data_combined)

# Création du graphique combiné avec la palette Set3
plt.figure(figsize=(12, 8))
sns.barplot(x='Graphe', y='Temps d\'inférence (ms)', hue='Nombre d\'évidences', palette='Set3', data=df_combined)
plt.title('Comparaison des temps d\'inférence avec une et deux évidences')
plt.show()

# Création du deuxième graphique combiné (Temps de propagation) avec la palette Set3
plt.figure(figsize=(12, 8))
sns.barplot(x='Graphe', y='Temps de propagation (s)', hue='Nombre d\'évidences', palette='Set3', data=df_combined)
plt.title('Comparaison des temps de propagation avec une et deux évidences')
plt.show()
```
