```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Donnes des deux tableaux combines
data_combined = {
    'Graphe': ['Polyarbre', 'Faiblement connects', 'Moyennement connect', 'Fortement connect',
               'Polyarbre', 'Faiblement connects', 'Moyennement connect', 'Fortement connect'],
    'Temps d\'infrence (ms)': [0, 27.459, 20.417, 33.943, 2, 2, 1, 2],
    'Temps de propagation (s)': [0, 0.027459, 0.020417, 0.033943, 5, 17, 14, 21],
    'Nombre d\'vidences': [1, 1, 1, 1, 2, 2, 2, 2]
}

# Cration du DataFrame
df_combined = pd.DataFrame(data_combined)

# Cration du graphique combin avec la palette Set3
plt.figure(figsize=(12, 8))
sns.barplot(x='Graphe', y='Temps d\'infrence (ms)', hue='Nombre d\'vidences', palette='Set3', data=df_combined)
plt.title('Comparaison des temps d\'infrence avec une et deux vidences')
plt.show()

# Cration du deuxime graphique combin (Temps de propagation) avec la palette Set3
plt.figure(figsize=(12, 8))
sns.barplot(x='Graphe', y='Temps de propagation (s)', hue='Nombre d\'vidences', palette='Set3', data=df_combined)
plt.title('Comparaison des temps de propagation avec une et deux vidences')
plt.show()
```
